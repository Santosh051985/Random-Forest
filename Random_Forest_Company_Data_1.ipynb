# The Model
The next part fits a random forest model to the data,

```python
X_train, X_test, y_train, y_test = train_test_split(comp_data.drop('nSal', 1), comp_data['nSal'], test_size = .2, random_state=10)
#split t
```

```python
model = RandomForestClassifier(n_jobs=2,oob_score=True,n_estimators=400,criterion="entropy",max_depth=5)
model.fit(X_train, y_train)
```

```python
estimator = model.estimators_[1]
feature_names = [i for i in X_train.columns]

y_train_str = y_train.astype('str')
y_train_str[y_train_str == '0'] = 'low sale'
y_train_str[y_train_str == '1'] = 'high sale'
y_train_str = y_train_str.values
```

```python
export_graphviz(estimator, out_file='tree.dot', 
                feature_names = feature_names,
                class_names = y_train_str,
                rounded = True, proportion = True, 
                label='root',
                precision = 2, filled = True)

from subprocess import call
call(['dot', '-Tpng', 'tree.dot', '-o', 'tree_sale.png', '-Gdpi=600'])

from IPython.display import Image
Image(filename = 'tree_sale.png')
```

This gives us on explainability tool. However, I can't glance at this and get a quick sense of the most important features of Company data set . We'll revisit those later. Next, let's evaluate the model,

```python
y_predict = model.predict(X_test)
y_pred_quant = model.predict_proba(X_test)[:, 1]
y_pred_bin = model.predict(X_test)
```

```python
confusion_matrix = confusion_matrix(y_test, y_pred_bin)
confusion_matrix
```

Sale evaluation  tests are often sold, marketed, cited and used with sensitivity and specificity as the headline metrics. 

```python
total=sum(sum(confusion_matrix))

sensitivity = confusion_matrix[0,0]/(confusion_matrix[0,0]+confusion_matrix[1,0])
print('Sensitivity : ', sensitivity )

specificity = confusion_matrix[1,1]/(confusion_matrix[1,1]+confusion_matrix[0,1])
print('Specificity : ', specificity)
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_quant)

fig, ax = plt.subplots()
ax.plot(fpr, tpr)
ax.plot([0, 1], [0, 1], transform=ax.transAxes, ls="--", c=".3")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.rcParams['font.size'] = 12
plt.title('ROC curve for diabetes classifier')
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')
plt.grid(True)
```

Another common metric is the Area Under the Curve, or AUC. This is a convenient way to capture the performance of a model in a single number, although it's not without certain issues. As a rule of thumb, an AUC can be classed as follows,
 - 0.90 - 1.00 = excellent
 - 80 - 0.90 = good
 - 0.70 - 0.80 = fair
 - .60 - 0.70 = poor
 - 0.50 - 0.60 = fail
Let's see what the above ROC gives us,

```python
auc(fpr, tpr)
```

# The Explanation 

Now let's see what the model gives us from the ML explainability tools.
Permutation importance * is the first tool for understanding a machine-learning model, and involves shuffling individual variables in the validation data (after a model has been fit), and seeing the effect on accuracy.

```python
perm = PermutationImportance(model, random_state=1).fit(X_test, y_test)
eli5.show_weights(perm, feature_names = X_test.columns.tolist())
```

So, it looks like the most important factors in terms of permutation is a helf Location at stores high impact on sales. The high importance of 'Price rate achieved' type makes sense,  also Shelf Location at storesas this is the immediate, subjective state of the price at the time of Sales (as opposed to, say, Urban_Yes,Education which is a much more general factor).
